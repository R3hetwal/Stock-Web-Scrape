```python
import requests
```

```python
html = requests.get(url).text
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
complete_data = []
for item in range(0, len(table)):
    
    country_data = []
    table_row = table[item]
    td = table_row.find_all('td')
    # print(td) # one table row has multiple table data
    for item in td:
        country_data.append(item.text)
         # extract data of each country
        print(country_data)
        
    complete_data.append(country_data)
    
```

```python
complete_data[0]
```

```python
complete_data[0][1]
```

```python
clean_data = list(map(lambda x: x[1:10], complete_data))
```

```python
for i in range(len(clean_data)):
  for j in range(len(clean_data[i])):
    if( clean_data[i][j] == ' ' or clean_data[i][j] == 'null' or clean_data[i][j] == 'NAN'):
       clean_data[i][j] = '100'
    else:   
      clean_data[i][j]=clean_data[i][j].replace('\n',' ')
      clean_data[i][j]=clean_data[i][j].replace('                                        ','')
```
